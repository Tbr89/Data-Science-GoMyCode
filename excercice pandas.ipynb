```python

data=data.assign(total_passengers=data['total_male']+data['total_female'] ) 
data




    










```

```python

data['travel_with'] = data.apply(lambda x: 'Alone' if pd.isnull(x['travel_with']) and x['total_passengers'] == 1 else x['travel_with'], axis=1)
```

```python
data.isnull().sum()
```

```python

data['travel_with'] = data.apply(lambda x: 'Spouse' if pd.isnull(x['travel_with']) and x['total_passengers'] == 2 else x['travel_with'], axis=1)
```

```python
correlation = data['total_male'].corr(data['total_cost'])
print(correlation)
```

```python
country_visits = data['country'].value_counts()
top_3_countries = country_visits.nlargest(3)
print(top_3_countries)
```

```python
#Determine the most common purpose of the visit by 'age_group'.
purpose_counts_by_age_group = data.groupby('age_group')['purpose'].value_counts()
most_common_purpose_by_age_group = purpose_counts_by_age_group.groupby(level=0).nlargest(1)
print(most_common_purpose_by_age_group)
```
